```python
import regex
```

```python
from bs4 import BeautifulSoup
```

```python
pattern = '([\p{Han}]+)' #抓漢字的正則運算法
```

```python
path = "asbc_a_001.xml" #这里是需要爬取的本地頁面
htmlfile = open(path, 'r', encoding='utf-8')
#開啟檔案
```

```python
soup = BeautifulSoup(htmlfile, 'xml') #这里是将content内容转化为BeautifulSoup格式的数据
arcticle_sentence = []
for sentence in soup.find_all('sentence'):
    arcticle_sentence.append(sentence)
#將article標籤裡的sentence標籤抓出來
```

```python
fp = open("asbc_a_001_write_sentence.txt", "w", encoding="utf-8")
```

```python
for sentence in arcticle_sentence:
    m = []
    m = regex.findall(pattern, str(sentence))#regex抓漢字
    for i in m:
        fp.write(i + " ")
    fp.write("\n\n")
#以sentence為單位將漢字抓出寫入
```

```python
fp.close
```
