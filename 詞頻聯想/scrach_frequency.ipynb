```python
import regex
```

```python
fp = open("修正低頻中心像表格.txt", "r" ,encoding="utf-8",errors="ignore")
line = fp.readline()
```

```python
pattern = '([\p{Han}]+)' #抓漢字的正則運算法
```

```python
data = {}
```

```python
line = fp.readline()
while(line):
    key = regex.findall(pattern, line) #抓詞頻答案
    print(key)
    line = fp.readline()
    line = fp.readline()
    line = fp.readline()
    ans = regex.findall(pattern, line) #抓詞頻題組
    print(ans)
    data[key[0]] = ans[2:]
    line = fp.readline()
```
