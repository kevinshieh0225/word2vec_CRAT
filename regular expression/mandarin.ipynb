```python
import regex
```

```python
fp = open("XXX.txt", "r" ,encoding="utf-8",errors="ignore")
line = fp.readline()
```

```python
pattern = '([\p{Han}]+)' #抓漢字的正則運算法
```

```python
m = regex.findall(pattern, line)#regex抓漢字
```

```python
str = ''
print(len(m))
print(type(m))
new = str.join(m)
#字串結合

print(new)
```
